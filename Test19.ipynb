```python
for column in df.columns:
    print(df[column].dtypes=='O')
```

```python
categorical = [column for column in df.columns if df[column].dtypes=='O' ]
```

```python
categorical
```

```python
numerical = [column for column in df.columns if df[column].dtypes!='O']
```

```python
numerical
```

```python
for variable in numerical:
    df[variable].hist(bins=10)
    plt.title("Distribution of {0}".format(variable) )
    plt.show()
```

```python
for variable in categorical:
#     x=df[variable].bar
    plt.title(variable)
    df[variable].value_counts().plot(kind='bar')
    plt.show()
```
